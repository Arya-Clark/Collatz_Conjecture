```python
# Collatz Conjecture Script

integer = int(input('input integer'))
counter = 1
print ('starting point: ',integer)

while integer > 1:
    if integer/2 == int(integer/2):
        integer = int(integer/2)
        counter = counter + 1
        print (integer)
        if integer == 1:
            break
    if integer/2 != int(integer/2):
        integer = 3*integer + 1
        counter = counter + 1
        print (integer)
        if integer == 1:
            break
    


print ('steps taken =', counter)


```
